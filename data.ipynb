```python
import pandas as pd 
import numpy as np
from scipy import stats

# visualization packages 
import matplotlib.pyplot as plt
import seaborn as sns

# for ignoring the warnings massages.
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
match = pd.read_csv('Match.csv')
match
```

```python
match.columns.values.tolist()
```

```python
for col in match.columns:
  match[col] = match[col].replace(np.nan, 0.0)
```

```python
match.away_team_goal
```

```python
match.home_team_goal
```

```python
match['winner'] = np.where(match.home_team_goal > match.away_team_goal, 1, 
         (np.where(match.away_team_goal > match.home_team_goal, 2, 0)))
```

```python
match.winner
```

```python
match.dtypes[match.dtypes != 'int64'][match.dtypes != 'float64']
```

```python
train = match.drop(['home_team_goal', 'away_team_goal', 'winner', 'season', 'date', 'goal', 
                    'shoton', 'shotoff', 'foulcommit', 'card', 'cross', 'corner', 'possession'], axis='columns')
test = match.winner
```

```python
test
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, test, test_size=0.2, random_state=8080)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier().fit(X_train, y_train)
pred = dt.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=8080)
lr.fit(X_train,y_train)
pred = lr.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```

```python
from sklearn.svm import SVC

svm = SVC(random_state=8080)
svm.fit(X_train,y_train)
pred = svm.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
pred = nb.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(random_state=8080)
mlp.fit(X_train,y_train)
pred = mlp.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
qda = QuadraticDiscriminantAnalysis()
qda.fit(X_train,y_train)
pred = qda.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```

```python
from sklearn.ensemble import AdaBoostClassifier
abc = AdaBoostClassifier(random_state=8080)
abc.fit(X_train,y_train)
pred = abc.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```

```python
from xgboost.sklearn import XGBClassifier
xgb = XGBClassifier(random_state=8080)
xgb.fit(X_train,y_train)
pred = xgb.predict(X_test)
print("Accuracy: " + str(accuracy_score(y_test, pred)))
```
